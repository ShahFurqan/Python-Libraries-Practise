```python
%pip install seaborn
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ourdata = pd.read_csv('Pokemon.csv')
```

```python
ourdata.head()
```

```python
sns.lmplot(x='Attack',y='Defense',data=ourdata,fit_reg=True) #fit_reg=True a line see in graph 
# if not want to see that so fit_reg=False                    # hue to colour our points
plt.show()
```

```python
sns.lmplot(x='Attack',y='Defense',data=ourdata,fit_reg=True,hue='Legendary') #fit_reg=True a line see in graph 
# if not want to see that so fit_reg=False
plt.show()
```

```python
ourdata['Legendary'].value_counts()
```

```python
fig = plt.figure()
a1=fig.add_axes([0,0,1,1]) # [0x,0y,width,height]

x= np.arange(1,10)
a1.plot(x,np.exp(x),'b')

a1.set_title('range of numbers')

#explicitly set x an y limits
plt.ylim(0,10000)
plt.xlim(0,10)

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```
