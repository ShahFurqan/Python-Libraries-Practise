# creating multiples plots

```python
classes = ['hohn','sanjay','marya','kashi','furqan','faizan','rehan']
scores = [10,12,12,15,16,13,19]

plt.figure(figsize=(10,5))

plt.subplot(231)          # 131 2 -->no of rows , 3 -->no columns, 1 --> position
plt.bar(classes,scores)

plt.subplot(232)
plt.scatter(classes,scores)

plt.subplot(233)
plt.plot(classes,scores)

plt.subplot(234)
plt.plot(classes,scores)

plt.suptitle('categorical plotting') # heading
plt.show()
```
