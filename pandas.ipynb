```python
sales = {'john':1000,
         'cri': 3000,
         'jakjs':121,
         'kkk': 1211,
         'lk':11
        }
region={'john':'west',
        'cri':np.nan,
        'jakjs':'wes',
        'kkk':np.nan,
        'lk':'north'
       }
```

# convefrting dict into DataFrame

```python
sales_df= pd.DataFrame.from_dict(sales,orient='index',columns=['sales'])
region_df= pd.DataFrame.from_dict(region,orient='index',columns=['region'])
```

```python
sales_df
```

# join

```python
joined_df=region_df.join(sales_df,how='left')      # those present in both that show others will be null
joined_df
```

# merge

```python
# here create common names of column

region_df.index.name='name'
sales_df.index.name='name'
```

```python
print(region_df)
print(sales_df)
```

```python
# now I merge on the names
print(pd.merge(region_df,sales_df,on='name',how='left'))   # just like join
```

```python
# now I merge on the names
print(pd.merge(region_df,sales_df,on='name',how='outer'))  # outer  # just like join 
```
